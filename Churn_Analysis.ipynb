# 📊 Key Insights

##1️⃣ Overall Churn Rate
Around 26–27% customers have churned.
Majority customers (about 73%) are retained.
This means churn is significant and needs action.
##2️⃣ Contract Type vs Churn
Month-to-month customers have the highest churn rate.
One-year and Two-year contract customers churn much less.
Longer contracts increase customer retention.
 ✅ Insight: Customers without long-term commitment are more likely to leave.
##3️⃣ Tenure vs Churn
Customers with low tenure (new customers) are more likely to churn.
Customers with higher tenure stay longer and are more loyal.
 ✅ Insight: The first few months are critical for retention.
##4️⃣ Monthly Charges vs Churn
Customers who churn generally have higher monthly charges.
High pricing may be a churn driver.
 ✅ Insight: Price sensitivity plays an important role.
##5️⃣ Payment Method vs Churn
Customers using Electronic check have the highest churn rate (~45%).
Customers using Automatic payment methods (bank transfer / credit card) have much lower churn.
 ✅ Insight: Manual payment users are more likely to churn.

# 📌 Final Conclusion

The analysis identifies contract type, tenure, monthly charges, and payment method as key drivers of churn. Customers with short-term contracts, lower tenure, higher monthly charges, and manual payment methods show higher churn risk. 

Focusing on long-term contracts, early customer engagement, and promoting automatic payments can significantly improve retention and reduce overall churn.
